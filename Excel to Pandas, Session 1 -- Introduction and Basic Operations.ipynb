# Accessing rows and columns:
We can access both rows and columns by position and by label. Accessing by labels is better (less prone to error, more understandable code)

Accessing by column labels required having a header. 

Accessing by row labels requires having an index. If no index is found, the position of the row is used as index





```python
# Selecting the "Open" column from the SP dataframe:

df_sp["Open"]
```

```python
# selecting a list of columns:

df_sp[["High", "Low", "Volume"]]
```

```python
# Selecting rows: no index is yet specified, so we use positions. 

# We specify multiple rows by slice operator or by a list. 

# Slice: We will select rows 20-30

df.loc[20:30]
```

```python
# List: select rows at positions 0, 20, 500, 1000, 3000

df.loc[[0,20,500,1000, 3000]]
```

```python
# But we don't need to use the position of rows. We have a natural index: the date. Let us set that as the index:

# NOTE: we have the option of modifying the dataframe inplace or creating a new version that has the index!
df_sp.set_index("Date", drop=True, inplace=True,verify_integrity=True)
df_sp.head()
```

```python
# Note that now "loc" function no longer accepts position indices, and needs dates:

df_sp.loc["1984-07-05"]
```

```python
# Just like before we can select mutiple rows using the slice or using a list:
# Select all rows for January 2005:

df_sp.loc["2005-01-01":"2005-01-31"]
```

```python
# We can also mix the selection of rows and columns. 

# Let's select only the high and low for SP dataframe, for only 2008:

df_sp.loc["2008-01-01":"2008-12-31", ["High", "Low"]]
```

```python
# Even after specifying an index, we can still select rows by
# position if needed. We will just need to use iloc, instead of loc:

df_sp.iloc[20:30]
```

```python
# and just like we can use a combination of row labels and column labels, we can use a combination of row positions and columns positions:
# Selecting columns 2-5, for rows 20-30

#(Note: Python indexing starts from 0. Column at index 2 is the 3rd column!)


df_sp.iloc[19:29, 1:4]
```

```python
# Similarily, we can get select slice of columns (if they are continuous) or a list of column positions:
df_sp.iloc[:, 2:5]
```

```python
# We can select by row positions and column positions at the same time:

df_sp.iloc[5000:5010, 2:4]
```

```python
# So far, we have just displayed the selected rows and columns. But we often we want to analyze them or do some operations on them

# The results from the selection operation are dataframes and can be stored in variables and operated on. Let's 
# Select the SP500 Open, Close, High, Low for 2008, and store it as a separate dataframe:

df_sp2008 = df_sp.loc["2008-01-01":"2008-12-31", ["Open", "Low", "High", "Close"]]
print(df_sp2008.shape)
print(df_sp2008.head())
```

```python
# This allows us to do the same summary statistics and all other dataframe
# operations on the selected rows/columns
df_sp2008.describe()
```

```python
# Side note: Columns can also be accessed as attributes of the dataframe:
# This only works if the column names are legal python identifier
#(no spaces, no punctuation other than _, etc.)
```

# Basic Plotting 

```python
# Let's visualize the closing prices of the S&P dataframe:

df_sp["Close"].plot()
```

```python
# Above Pandas automatically used the Date for the y-axis.
# This only works because we have already set the date column as the index.

# Here is what happens when we plot a Series from the Dax dataframe where
# the index has not been set yet:

df_dax["Close"].plot()
```

```python
# We can plot two lines, and by default 
# they will appear in the same panel.
# Let's plot S&P closing and the Nikkei closing prices

# Note: first, we need to set the date as index for the other df

df_nikkei.set_index("Date", inplace = True)
df_dax.set_index("Date", inplace=True)
df_dax["Close"].plot()
df_nikkei["Close"].plot()
```

```python
# We can add a title and a legend. We can also change the colors
# by default the legend is the name of the columns

df_dax["Close"].plot(title="Comparing performace of DAX to Nikkei",
                     legend = True,
                    color = "red")
df_nikkei["Close"].plot( legend="True", color = "green")
```

```python
# Note: there are several ways to modify the legend if the column names
# are not descriptive enough. Below is one example: 

fig, ax = matplotlib.pyplot.subplots()
df_dax["Close"].plot(title="Comparing performace of DAX to Nikkei", legend = True)
df_nikkei["Close"].plot(legend=True)
ax.legend(["DAX", "NIKKEI"])

# But still, the simplest way to do this is to rename the columns as
# approperite. We will see next how to re-name columns
```

# Adding and Deleting Columns

```python
# We can create a new a column in several ways. For example,
# by operating on existing columns.

# So we can create a column for the change between open and close. 

df_sp2008["Close"] - df_sp2008["Open"]
```

```python
# We will add this a new Column (Series) to the dataframe, let's name it
# "Change_from_open"
df_sp2008["Change_from_open"] = df_sp2008["Close"] - df_sp2008["Open"]
df_sp2008.head()
```

```python
# Deleting a column from a dataframe is easy, just use drop
# we have to remember to specify that we are dropping a column not row
# by specifying axis = 1
# We also can specify whether we want the drop applied to the
# current df, or whether we want a new modified df
df_sp2008.drop("Low", axis = 1, inplace= True)
df_sp2008.head()
```

# Saving DataFrames to an Excel Spreadsheet

```python
# NOTE: this is meant to save spreadsheets to a new excel workbook
# if an existing file is given to the function, that file will be overwritten!

# LET's save the sp2008 spreadsheet to a new workbook

# The process of saving 1 spreadsheet is simple:

df_sp2008.to_excel("SP2008.xlsx", sheet_name="sp2008_processed")
```

```python
# we can save more than 1 spreadsheet to a workbook.

# This requires creating a writer object, writing each sheet to the file
# then saving the workbook 

writer = pd.ExcelWriter("workbook2.xlsx")
df_sp2008.to_excel(writer, sheet_name="SP500_2008")
df_dax.to_excel(writer, sheet_name= "Dax30")
writer.save()
```
